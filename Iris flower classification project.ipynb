```python
#ML BC iris flower classification project
import pandas as pd
import numpy as np
dataset=pd.read_csv('iris flower dataset.csv')
dataset
```

```python
#separating independent variables from dependent
x=dataset.iloc[:,0:-1].values
x
#y=dataset.iloc[:,-1].values
#y
#converting labels into integers(encoding)
#from sklearn.preprocessing import LabelBinarizer
#le=LabelBinarizer()
```

```python
y=dataset.iloc[:,-1].values#dependent variable
y
```

```python
#ML model understands only numerical values
#eg:male and female
#Male-0 Female-1-->these are called labels
#One hot encoding
#here labels are made into columns
#1  2
#1  0
#0  1
#incase of this project,setosa-1,versicolor-2,virginica-3
#1  2  3
#1  0  0 here setosa is 1 rest all are zero..same applies to rest of them
#0  1  0
#0  0  1
#this can be done by both one hot encoding and label encoding
from sklearn.preprocessing import LabelBinarizer
#importing library
```

```python
le=LabelBinarizer()
y=le.fit_transform(y)
y
```

```python
#dividing dataset into train and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
x_train


```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#creating a classifier object that will train the model based on
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
classifier.fit(x_train,y_train)
```

```python
#predicting using test sets
y_pred=classifier.predict(x_test)
y_pred
```

```python
#metrics from sklearn library-->used to calculate the accuracy
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_test.argmax(axis=1),y_pred.argmax(axis=1))#if we have labels(here it is setosa,virginica..)(3 labels here) which is converted to 3*3 matrix where we have values crt values in diagnol only.If we have value in other place other than diagnol it has some error,if only the diagnol elements has some value and rest of them is zero means, our model has more accuracy.
cm
#acc=accuracy_score(y_test.argmax(axis=1),y_pred.argmax(axis=1))
```

```python
#In this case we are having 98% accuracy.
acc=accuracy_score(y_test.argmax(axis=1),y_pred.argmax(axis=1))
acc
```
